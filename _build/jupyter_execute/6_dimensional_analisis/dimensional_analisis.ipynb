# Análisis dimensional y semejanza

## Introducción

En mecánica de fluidos rara vez podemos ensayar directamente el prototipo final. Un avión, una tubería o una bomba suelen estudiarse primero mediante modelos, correlaciones o datos de laboratorio. El análisis dimensional entrega el lenguaje para conectar esos ensayos con el sistema real.

Consideremos que la fuerza de arrastre sobre una esfera de diámetro $D$. 

<img src="./figures/figure1_drag_sphere.png" width="400px" align= center>

<center> <small><b>Figura 1.</b> Fotografía del arrastre en una esfera sumergida en agua. <i>Van Dyke, An Album of Fluid Motion (1982)</i> </small></center>

Mediante un análisis fenomenológico, se puede demostrar que esta fuerza depende de la densidad del fluido ($\rho$), la viscosidad ($\mu$), la velocidad ($V$) y una longitud característica ($L_c = D$ en este caso):

\begin{equation*}
F = f(\rho,\mu,V,L_c)
\end{equation*}


Debido a la gran cantidad de parámetros, si quisieramos elaborar un estudio variando cada uno por separado, el análisis del problema se vuelve costoso y dificil de interpretar. 

La idea del análisis dimensional es reorganizar el fenómeno para encontrar una descripción más compacta y útil. Para ello, se identifican las variables relevantes y se agrupan en términos adimensionales, lo que permite simplificar el problema y revelar relaciones fundamentales entre ellas.

Por ejemplo, en el caso de la fuerza de arrastre, podemos demostrar que los cuátro parámetros pueden agruparse en una función de la forma:

\begin{equation*}
\frac{F}{\rho V^2 D^2} = f\left(\frac{\rho V D}{\mu}\right)
\end{equation*}

## Principio de homogeneidad dimensional

### Dimensiones y unidades

El análisis dimensional trabaja con dimensiones, no con unidades particulares. 
- Una **unidad** es una convención de medida, como $\mathrm{m}$, $\mathrm{s}$ o $\mathrm{kg}$
- Una **dimensión** expresa la naturaleza física de una variable, como longitud $L$, masa $M$ o tiempo $T$.


Ejemplos de dimensiones de variables comunes en mecánica de fluidos son:

\begin{align*}
[V] &= LT^{-1} \\
[\rho] &= ML^{-3} \\
[\mu] &= ML^{-1}T^{-1} \\
[p] &= ML^{-1}T^{-2}
\end{align*}


>**Nota.** Dos ecuaciones pueden estar escritas en unidades distintas y seguir representando la misma física, siempre que conserven la misma estructura dimensional.

### Homogeneidad dimensional

El principio de homogeneidad dimensional establece: ***una ecuación físicamente válida debe tener la misma dimensión en todos sus términos aditivos.*** En palabras simples, no se pueden sumar magnitudes de distinta naturaleza física.


Por ejemplo, la ecuación de movimiento de un cuerpo bajo la acción de la gravedad:

\begin{equation*}
x(t) = x_0 + V_0 t + \frac{1}{2}gt^2
\end{equation*}

es homogénea porque todos los términos tienen dimensión (longitud, $L$)

Cuando una relación es dimensionalmente homogénea, muchas veces puede reescribirse en forma adimensional, como en el caso de la fuerza de arrastre:

\begin{equation*}
\frac{F}{\rho V^2 L^2} = \phi\left(\frac{\rho V L}{\mu}\right)
\end{equation*}

Ahora el fenómeno queda descrito con dos cantidades adimensionales en vez de cuatro variables dimensionales independientes.

## Teorema de Buckingham Pi




***Si un fenómeno depende de $n$ variables dimensionales y esas variables involucran $r$ dimensiones fundamentales independientes, entonces la relación física puede expresarse mediante $n-r$ grupos adimensionales independientes.***



Ese resultado no reemplaza el juicio físico: primero hay que identificar correctamente qué variables son relevantes.

**Procedimiento práctico.**

1. Listar las variables que intervienen en el problema.
2. Elegir la variable de salida que interesa predecir.
3. Escribir las dimensiones de cada variable.
4. Seleccionar **variables repetitivas** que cubran las dimensiones fundamentales.
5. Formar grupos adimensionales y reconocer los que tienen significado físico conocido.


**¿Cómo elegir variables repetitivas?**

- En conjunto deben contener todas las dimensiones fundamentales del problema.
- No deben formar por sí solas un grupo adimensional.
- No conviene incluir entre ellas la variable dependientes que se quiere correlacionar.

Además, es preferible que representen geometría, cinemática y propiedades del fluido de una forma físicamente clara.


### Ejemplo 1: Fuerza de arrastre

Para un cuerpo de forma fija en un flujo incompresible, sin superficie libre y a baja velocidad, planteamos:

\begin{equation*}
F = f(\rho,\mu,V,D)
\end{equation*}

1. **Determinar el número de grupos adimensionales**: Aquí hay $n=5$ variables si contamos a $F$, y aparecen $r=3$ dimensiones fundamentales. Por lo tanto, esperamos $5-3=2$ grupos adimensionales que denotaremos como $\Pi_1$ y $\Pi_2$.

2. **Elegir variables repetitivas**: Para cubrir las dimensiones fundamentales de masa, longitud y tiempo, podemos elegir a $\rho$, $V$ y $D$ como variables repetitivas. Estas variables representan propiedades del fluido, cinemática y geometría respectivamente, lo que facilita la interpretación física de los grupos adimensionales resultantes.

3. **Formar grupos adimensionales**: Utilizando las variables repetitivas, determinamos la combinación que permite adimensionalizar las variables restantes, $F$ y $\mu$.

En forma matemática, queremos determinar los exponentes $a_i$, $b_i$ y $c_i$ ($i = 1, 2$) asociados a $[\rho]$, $[V]$ y $[D]$ para formar los grupos adimensionales $\Pi_1$ y $\Pi_2$:

\begin{align*}
\Pi_1 = \rho^{a_1} V^{b_1} D^{c_1} F  
\Rightarrow 
\left(\frac{M}{L^3}\right)^{a_1} \left(\frac{L}{T}\right)^{b_1} \left(\frac{L}{T}\right)^{c_1} \left(\frac{ML}{T^2}\right) &= M^0L^0T^0
\\
(M)^{a_1+1}(L)^{-3a_1+b_1+c_1+1}(T)^{-b_1-c_1-2} &= M^0L^0T^0
\end{align*}

\begin{equation*}
\begin{aligned}
&\mathrm{M}:&\quad a_1 + 1 = 0 \\
&\mathrm{L}:&\quad -3a_1 + b_1 + c_1 + 1 = 0 \\
&\mathrm{T}:&\quad -b_1 - c_1 - 2 = 0
\end{aligned}
\quad\Rightarrow\quad
\begin{aligned}
a_1 &= -1 \\
b_1 &= -2 \\
c_1 &= -2
\end{aligned}
\end{equation*}

\begin{equation*}
\Pi_1 = \frac{F}{\rho V^2 D^2}
\end{equation*}

Igualmente, para $\Pi_2$

\begin{align*}
\Pi_2 = [\rho]^{a_2} [V]^{b_2} [D]^{c_2} [\mu] 
\Rightarrow 
\left(\frac{M}{L^3}\right)^{a_2} \left(\frac{L}{T}\right)^{b_2} \left(\frac{L}{T}\right)^{c_2} \left(\frac{ML}{T}\right) &= M^0L^0T^0
\\
(M)^{a_2+1}(L)^{-3a_2+b_2+c_2+1}(T)^{-b_2-c_2-1} &= M^0L^0T^0
\end{align*}

\begin{equation*}
\begin{aligned}
&\mathrm{M}:&\quad a_2 + 1 = 0 \\
&\mathrm{L}:&\quad -3a_2 + b_2 + c_2 + 1 = 0 \\
&\mathrm{T}:&\quad -b_2 - c_2 - 1 = 0
\end{aligned}
\quad\Rightarrow\quad
\begin{aligned}
a_1 &= -1 \\
b_1 &= -1 \\
c_1 &= -1
\end{aligned}
\end{equation*}

\begin{equation*}
\Pi_2 = \frac{\rho V D}{\mu}
\end{equation*}

4. **Definir funcional adimensional** La relación adimensional puede escribirse como

    \begin{equation*}
    C_D = \phi(\mathrm{Re})
    \end{equation*}
    
    donde $C_D = \frac{F}{\rho V^2 D^2}$ es el coeficiente de arrastre, y $\mathrm{Re} = \frac{\rho V D}{\mu}$ es el número de Reynolds.

El nuevo funcional adimensional ofrece una simplifación del problema, ya que ahora la fuerza de arrastre se describe en términos de dos cantidades adimensionales, facilitando el análisis y la interpretación de resultados experimentales. Por otro lado, la formulación ofrece una interpretación física más clara: *cómo un régimen de flujo (caracterizado por el número de Reynolds) afecta el arrastre experimentada por el cuerpo.*

### Ejemplo 2: caída de presión en tubería

Para una tubería recta de diámetro $D$, longitud $L_0$ y rugosidad $\varepsilon$, una formulación general es:

\begin{equation*}
\Delta p = f(\rho,\mu,V,D,L_0,\varepsilon)
\end{equation*}


Siguiendo los pasos del ejemplo anterior tenemos:

1. **Número de grupos adimensionales:** $n=7$ variables y $r=3$ dimensiones fundamentales, por lo que esperamos $7-3=4$ grupos adimensionales.


2. **Variables repetitivas:** Seleccionamos nuevamente $\rho$, $V$ y $D$


3. **Formar grupos adimensionales**: Utilizando las variables repetitivas, determinamos la combinación que permite adimensionalizar las variables restantes, $\Delta p$, $\mu$, $L_0$ y $\varepsilon$.
\begin{equation*}
\Pi_1 = \frac{\Delta p}{\rho V^2}, \quad
\Pi_2 = \frac{\mu}{\rho V D}, \quad
\Pi_3 = \frac{L_0}{D}, \quad
\Pi_4 = \frac{\varepsilon}{D}
\end{equation*}


4. **Definir funcional adimensional** La relación adimensionalpuede escribirse como
    
    \begin{equation*}
    \frac{\Delta p}{\rho V^2} = \phi\left(\mathrm{Re},\frac{L_0}{D},\frac{\varepsilon}{D}\right)
    \end{equation*}

    Esto anticipa por qué en flujo interno las pérdidas no dependen solo de $\mathrm{Re}$, sino también de la escala geométrica y de la rugosidad relativa.

## Definición de números adimensionales característicos

Un número adimensional no es solo una abreviatura algebraica. En general expresa la competencia entre efectos físicos.


### Número de Reynolds: inercia frente a viscosidad
Para una longitud característica $L_c$ (el diámetro, por ejemplo), el número de Reynolds se define como

\begin{equation}
\mathrm{Re} = \frac{\rho V L_c}{\mu}
\end{equation}

Compara efectos de inercia ($\propto\rho V^2$) con esfuerzos viscosos ($\propto \mu V/L_c$).

- Si $\mathrm{Re}$ es pequeño, la viscosidad domina y amortigua el movimiento.
- Si $\mathrm{Re}$ es grande, la inercia domina en gran parte del flujo, aunque la viscosidad sigue siendo importante cerca de paredes y capas de corte.


### Número de Froude: inercia frente a gravedad

\begin{equation}
\mathrm{Fr} = \frac{V}{\sqrt{gL_c}}
\end{equation}

Compara fuerzas de inercia ($\propto\rho V^2$) con fuerzas de gravedad ($\propto \rho g L_c$).

Es el parámetro dominante en flujos con superficie libre, como canales, vertederos, aliviaderos, oleaje y resistencia de buques.


### Número de Euler: presión frente a inercia

\begin{equation}
\mathrm{Eu} = \frac{\Delta p}{\rho V^2}
\end{equation}

El número de Euler ($\mathrm{Eu}$) se utiliza, principalmente, en flujos con grandes caidas de presión $\Delta p$. Particularmente, en problemas de cavitación, $\Delta p = p_a - p_v$, donde $p_a$ es la presión absoluta y $p_v$ es la presión de vapor del fluido.

### Número de Mach: compresibilidad

\begin{equation}
\mathrm{Ma} = \frac{V}{a}
\end{equation}

El número de Mach ($\mathrm{Ma}$) compara la velocidad del flujo con la velocidad del sonido y marca cuándo la compresibilidad deja de ser despreciable.




Como regla práctica, cuando $\mathrm{Ma} \gtrsim 0.3$, ya conviene vigilar efectos compresibles.

## Adimensionalización de las ecuaciones básicas

El analisis dimensional también tiene uso práctico para el análisis de ecuaciones de conservación. En este caso, el objetivo es reescribir las ecuaciones en forma adimensional escogiendo escalas de referencia para longitud, velocidad, tiempo y presión. La estrategia pone en evidencia qué parámetros gobiernan la física y en qué términos aparecen.


- Para un flujo incompresible con viscosidad constante, una elección natural es usar una longitud característica $L$ y una velocidad característica $U$.

- Usamos estas escalas para redinfinir las variables adimensionales de las ecuaciones de conservación:

    \begin{align*}
    x^* &= \frac{x}{L}, & y^* &= \frac{y}{L}, & z^* &= \frac{z}{L}, \\
    t^* &= \frac{tU}{L}, & \vec{V}^* &= \frac{\vec{V}}{U}, & p^* &= \frac{p}{\Delta p}
    \end{align*}

    >**Nota.** $\Delta p$ representa una escala de presión dependiendo del problema (gradiente de presión, por ejemplo). También puede ser sustituido por una presión constante (presión atmosférica, por ejemplo), o por la presión dinámica $\frac{1}{2}\rho U^2$.

- Al sustituir en continuidad y Navier-Stokes, las ecuaciones toman una forma mucho más informativa:

    \begin{align*}
    \nabla^*\cdot\vec{V}^* &= 0 \\
    \frac{\mathrm{D}\vec{V}^*}{\mathrm{D}t^*} &= -\frac{ 1}{\text{Eu}}\nabla^* p^* + \frac{1}{\text{Fr}^2}\hat{g} + \frac{1}{\mathrm{Re}}\nabla^{*2}\vec{V}^*
    \end{align*}

La expresión final indica los fenómenos que dominan el flujo. Dado que los operadores matemáticos son adimensionales ($\nabla^*$, $\nabla^{*2}$, etc. $\in [0, 1]$), la competencia entre efectos físicos queda explícita a través de los números adimensionales que aparecen como coeficientes de cada término.


### Criterios de semejanza según el tipo de flujo

- Flujo externo incompresible y sin superficie libre: típicamente domina $\mathrm{Re}$.
- Flujo con superficie libre: suele dominar $\mathrm{Fr}$; $\mathrm{Re}$ y, a veces, $\mathrm{We}$ también pueden importar.
- Flujo compresible: suelen importar $\mathrm{Ma}$ y $\mathrm{Re}$.
- Flujo interno en tuberías: aparecen $\mathrm{Re}$, $\varepsilon/D$ y con frecuencia $L/D$.

La física decide qué grupos son relevantes; el método solo los organiza.


## Modelación y semejanza

El objetivo de un modelo no es solo entregar resultados. El objetivo es que esos resultados representen al prototipo mediante una ley de escala bien fundamentada. En este contexto, ***un modelo y un prototipo presentan semejanza completa cuando existe semejanza geométrica y, además, los grupos adimensionales independientes que gobiernan el fenómeno tienen los mismos valores en ambos sistemas.***  Estas condiciones, además aseguran semejanza cinemática y dinámica: los patrones de movimiento son homólogos y las fuerzas correspondientes mantienen las mismas razones relativas


En la práctica, sin embargo, no siempre es posible lograr semejanza completa. En esos casos, se busca una semejanza parcial que capture los fenómenos más relevantes para el problema en cuestión. Por ejemplo, en un túnel de viento a baja velocidad, es posible lograr semejanza en número de Reynolds pero no en número de Mach, lo que es aceptable si el flujo es incompresible.


### Semejanza geométrica

Existe **semejanza geométrica si todas las longitudes homólogas conservan la misma razón de escala**:

\begin{equation*}
\lambda_L = \frac{L_m}{L_p} = \frac{x_m}{x_p} = \frac{y_m}{y_p} = \frac{z_m}{z_p}
\end{equation*}


Eso incluye forma, posiciones relativas y detalles geométricos que afecten el flujo. Particularmente, en el caso de ángulos, **la semejanza geométrica implica que los ángulos homólogos deben ser iguales.**

### Semejanza cinemática

Existe semejanza cinemática cuando, además de la geometría, los campos de velocidad y las trayectorias homólogas mantienen una escala temporal consistente.

\begin{equation*}
\lambda_V = \frac{V_m}{V_p},\quad
\lambda_T = \frac{T_m}{T_p}, \quad
\lambda_V = \frac{\lambda_L}{\lambda_T}
\end{equation*}

Formalmente, la definición se expresa como: ***Los movimientos de dos sistemas son cinemáticamenle semejantes si partículas homólogas alcanzan puntos homólogos en instantes homólogos.***


<img src="./figures/figura4_semenjanza_cine.png" width="700px" align= center>

<center> <small><b>Figura 4.</b> Semejanza cinemática en flujo sin superficie libre: las velocidades homólogas del modelo y del prototipo conservan una razón de escala común.</small></center>

La semejanza cinemática suele ser suficiente para flujos sin fricción, como en aerodinámica externa a alta velocidad. Sin embargo, en flujos con fricción, como en tuberías o alrededor de cuerpos, la semejanza cinemática no garantiza que los esfuerzos de corte homólogos sean iguales, lo que puede afectar la validez de las predicciones.

### Semejanza dinámica

Existe semejanza dinámica cuando las razones entre fuerzas correspondientes son equivalentes en modelo y prototipo.
