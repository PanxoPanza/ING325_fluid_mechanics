## VC fijo con variación de momentum al interior: llenado de una cinta transportadora

En esta sección resolveremos el problema de una cinta transportadora horizontal que se mueve a velocidad constante y que recibe un flujo de arena desde una tolva. La formulación es un ejemplo clásico para aplicar la ecuación de conservación del momento lineal en un volumen de control fijo con acumulación de masa en su interior. El enunciado, tomado del libro de Fox & McDonald y reproducido en el material de curso, se resume a continuación.

### Descripción del problema

Se tiene una cinta transportadora horizontal que se desplaza con velocidad constante de $1,\text{m/s}$. Desde una tolva cae arena verticalmente hacia la cinta con una velocidad relativa de $1{,}5,\text{m/s}$ y un flujo másico constante de $\dot m=230,\text{kg/s}$ (la densidad de la arena es aproximadamente $1602,\text{kg/m}^3$). La cinta está inicialmente vacía, pero se va llenando de arena con el tiempo. Si se desprecia la fricción en el sistema de accionamiento y en los rodillos, ¿cuál es la tensión necesaria para tirar de la cinta mientras ésta se está llenando?

### Selección del volumen de control y supuestos

Para aplicar la ecuación de conservación del momento lineal elegimos un volumen de control fijo que engloba una porción de la cinta entre dos secciones verticales (1 y 2) y que se extiende desde la superficie de la cinta hasta el punto de descarga de la tolva. Este volumen de control incluye la arena que cae y la parte de la cinta donde se acumula la arena (véase el rectángulo punteado en la figura). Los supuestos principales son:

- La cinta es horizontal, flexible y se mueve a una velocidad constante $V_b=1,\text{m/s}$.
- La arena es incompresible y cae verticalmente con velocidad $1{,}5,\text{m/s}$ respecto al suelo; esto implica que no tiene componente horizontal al ingresar al volumen de control.
- La fricción en los rodillos y en el mecanismo de accionamiento es despreciable.
- Se desprecia el peso de la cinta y de la arena en el análisis horizontal (el peso actúa en la dirección vertical y no afecta la componente horizontal del momento lineal).

### Conservación de masa

Dentro del volumen de control, la arena se acumula sobre la cinta. Sea $m(t)$ la masa de arena contenida en el volumen de control. La ecuación de conservación de masa en forma integral indica

\begin{equation}
\frac{\mathrm d m(t)}{\mathrm dt} = \dot m_{\text{entra}} - \dot m_{\text{sale}},
\end{equation}

donde $\dot m_{\text{entra}}$ es el flujo másico de arena que entra desde la tolva ($230,\text{kg/s}$) y $\dot m_{\text{sale}}$ es el flujo másico de arena que sale del volumen de control junto con la cinta. Al inicio del llenado $\dot m_{\text{sale}}$ puede ser menor que $\dot m_{\text{entra}}$ porque la arena se acumula en la cinta. Sin embargo, como veremos a continuación, la solución final no depende de esta diferencia.

### Conservación de momento lineal

La ecuación de conservación del momento lineal en dirección horizontal para un volumen de control fijo se expresa como

\begin{equation}
\sum F_x^{,\text{ext}} = \frac{\mathrm d}{\mathrm dt}\int_{\text{VC}} \rho u,\mathrm dV + \sum_{\text{salidas}} \dot m, u - \sum_{\text{entradas}} \dot m, u,
\end{equation}

donde $u$ es la componente horizontal de la velocidad absoluta del material (arena o cinta) y $\sum F_x^{,\text{ext}}$ es la suma de fuerzas externas en la dirección horizontal.

### Cálculo de las contribuciones
Tasa de variación del momento dentro del VC. La arena acumulada dentro del volumen de control se mueve con la misma velocidad horizontal que la cinta, $u=V_b$. Por consiguiente, el momento lineal horizontal en el interior es $m(t),V_b$ y su derivada temporal es
\begin{equation}
\frac{\mathrm d}{\mathrm dt}\bigl[ m(t),V_b \bigr] = \frac{\mathrm d m}{\mathrm dt},V_b = \bigl(\dot m_{\text{entra}} - \dot m_{\text{sale}}\bigr) V_b.
\end{equation}
Esta contribución se denomina término no estacionario y representa la aceleración de la masa almacenada.
Flujo de momento en las entradas y salidas. La arena que entra desde la tolva no posee componente horizontal de velocidad, de modo que no aporta flujo de momento lineal en dirección $x$ en la entrada. En la salida, la arena abandona el volumen de control sobre la cinta con velocidad horizontal $V_b$, por lo que la contribución de salida es $\dot m_{\text{sale}},V_b$.
Fuerzas externas en dirección horizontal. La única fuerza horizontal significativa es la debida a la tensión de la cinta. Denotaremos por $T_1$ la tensión en la cinta en la sección 1 y por $T_2$ la tensión en la sección 2. En la sección 1 (izquierda) la cinta es halada hacia la derecha por la tensión $T_1$, de manera que la cinta ejerce sobre el volumen de control una fuerza horizontal $+T_1$. En la sección 2 (derecha) la cinta es halada hacia la derecha por la tensión $T_2$, lo que significa que ejerce sobre el volumen de control una fuerza horizontal $-T_2$ (hacia la izquierda). Por lo tanto, la suma de fuerzas externas en dirección $x$ es
\begin{equation}
\sum F_x^{,\text{ext}} = T_1 - T_2.
\end{equation}
### Ecuación de momento lineal en la dirección horizontal

Sustituyendo las contribuciones anteriores en la ecuación general obtenemos

\begin{align}
T_1 - T_2 &= \underbrace{\bigl(\dot m_{\text{entra}} - \dot m_{\text{sale}}\bigr) V_b}{\text{término no estacionario}}
+ \underbrace{\dot m{\text{sale}},V_b}{\text{flujo de salida}} - \underbrace{\dot m{\text{entra}},(0)}{\text{flujo de entrada}} \
&= \dot m{\text{entra}}, V_b.
\end{align}

Sorprendentemente, el resultado final no depende del flujo de salida ni de la tasa de acumulación, pues ambos términos se combinan de manera que siempre aparece el flujo másico de entrada multiplicado por la velocidad de la cinta. Así, la diferencia de tensiones entre las secciones 1 y 2 (también denominada la “fuerza de arrastre” requerida por el motor) viene dada por

\begin{equation}
\Delta T = T_1 - T_2 = \dot m_{\text{entra}}, V_b.
\end{equation}

### Cálculo numérico de la tensión

Con los datos del problema:

Flujo másico de arena: $\dot m_{\text{entra}} = 230,\text{kg/s}$.
Velocidad horizontal de la cinta: $V_b = 1,\text{m/s}$.

La diferencia de tensiones necesaria para mantener la cinta en movimiento es

\begin{equation}
\Delta T = (230,\text{kg/s}),(1,\text{m/s}) = 230,\text{N}.
\end{equation}

Interpretación física: cada segundo caen 230 kg de arena sin velocidad horizontal. Para que esta masa se desplace con la cinta a $1,\text{m/s}$ se requiere un impulso horizontal de $230,\text{kg}\cdot\text{m/s}$, lo que equivale a una fuerza constante de $230,\text{N}$. Este es el diferencial de tensión que debe proporcionar el accionamiento entre las ramas de la cinta. En una cinta sin fricción las tensiones en ambas ramas suelen ser aproximadamente iguales, por lo que la tensión en cada rama es prácticamente $T_1 \approx T_2 + 230,\text{N}$. El motor debe suministrar esa diferencia de fuerzas para compensar la aceleración de la arena.

### Conclusiones
El análisis se realizó sobre un volumen de control fijo que incluye la zona donde la arena cae y se acumula. Aunque la masa de arena dentro del volumen de control aumenta con el tiempo, el balance de momento lineal muestra que la fuerza horizontal necesaria depende únicamente del flujo másico entrante y de la velocidad de la cinta.
La acumulación de arena sólo aparece en el término no estacionario, pero se cancela al combinarse con el flujo de salida. En consecuencia, el resultado final es el mismo tanto al inicio del llenado como en régimen estacionario.
Para los datos proporcionados, el accionamiento debe generar una diferencia de tensión de aproximadamente $230,\text{N}$ para mantener la cinta a velocidad constante mientras se llena.

Este ejemplo ilustra cómo el teorema de transporte de Reynolds y la conservación del momento lineal permiten analizar fenómenos transitorios de manera sencilla, proporcionando una herramienta útil para ingenieros en formación.
