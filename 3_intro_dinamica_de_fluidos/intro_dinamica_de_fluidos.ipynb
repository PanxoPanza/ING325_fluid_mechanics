# Introducción a la dinámica de fluidos

## Sistema y volúmen de control

### Sistema

Definimos como **sistema** a una *cantidad de materia de masa fija elegida para el estudio*

En general, para el estudio de **sólidos**, la definición de sistema es sencilla, ya que este considera la interacción del cuerpo con las fuerzas externas.

Sin embargo, en el caso de **fluidos**, la definición no es tan sensilla ya que el sistema ocupa un espacio infinito.

### Volumen de control

Para estudiar dinámica de fluidos utilizamos el concepto de **volumen de control**, una región imaginaria en el espacio donde aplicaremos nuestro análisis.

La forma y comportamiento de las fronteras del volumen de control son ajustadas a cada problema en particular. A medida que vayamos revisando problemas, entenderemos cuál es la forma más conveniente para definirlas.

<img src="./figures/control_volume.png" width="700px" align= center>

>**Nota.** Las frontera de un volumen de control pueden ser permeables o impermeables, móbiles o fijas.

## Teorema de transporte de Reynolds (TTR)

El **teorema de transporte de Reynolds** es un teorema fundamental que permite reformular las leyes de la física de sistemas para un volúmen de control.

### Leyes de conservación aplicadas a un sistema

Definimos como **sistema** al conjunto de objetos o entidades materiales que están relacionados entre sí. Ejemplos de sistemas son un motor, un planeta, una molécula, un bloque sólido, etc. Cada sistema debe satisfacer las denominadas leyes de conservación.

<img src="./figures/sistema_ejemplo.png" width="300px" align= center>

<center><small>Ejemplo de dos sistemas A y B asociados por una polea y una cuerda.</small></center>

Consideremos algunas **leyes de conservación fundamentales** aplicadas a un sistema:

- Masa ($m_\mathrm{sys}$), 

$$\frac{d}{dt}m_\mathrm{sys} = 0$$

- Momento lineal $(m\vec{V})_\mathrm{sys}$, 

$$\frac{d}{dt}(m\vec{V})_\mathrm{sys} = \vec{F}_\mathrm{neta}$$

- Momento angular $\vec{L}_\mathrm{sys}=(\vec{r}\times m\vec{V})_\mathrm{sys}$, 

$$\frac{d}{dt}\vec{L}_\mathrm{sys} = \vec{M}_\mathrm{neto}$$

- Energía $E_\mathrm{sys} =(m\tilde{e})_\mathrm{sys}$, 

$$\frac{d}{dt}(m\tilde{e})_\mathrm{sys} = \dot{Q} - \dot{W}$$

donde $\vec{F}_\mathrm{neta}$ y $\vec{M}_\mathrm{neto}$ son, respectivamente, *la fuerza y torque aplicado sobre el sistema*, $\tilde{e}$ es la *energía específica (energía por unidad de masa)*, $\dot{Q}$ es la *tasa de transferencia de calor*, y $\dot{W}$ es *potencia*.

En cada ley de conservación notamos que hay una ***propiedad extensiva $B_\mathrm{sys}$ o neta***  (por ejemplo, $E_\mathrm{sys}$), y una ***propiedad intensiva $\beta$ por unidad de masa*** (por ejemplo, $\tilde{e}_\mathrm{sys} = E_\mathrm{sys}/m_\mathrm{sys}$).

### Formulación general

El **teorema de transporte de Reynolds** establece una relación entre la **variación temporal de una propiedad extensiva del sistema ($B_\mathrm{sys}$)** y su respectiva **propiedad intensiva dentro del volumen de control ($\beta$)**

En su forma más general, para un **volumen de control móvil y deformable**:

\begin{equation}
\frac{d}{dt}B_\mathrm{sys} = \frac{d}{dt}\int_\mathrm{VC} \rho\beta d\forall + \int_\mathrm{VC} \rho\beta\vec{V}_r\cdot\hat{n}dA
\end{equation}

donde: 
- $\displaystyle \frac{d}{dt}B_\mathrm{sys}$ es la **tasa de cambio de $B_\mathrm{sys}$ en el sistema.**

- $\displaystyle \frac{d}{dt}\int_\mathrm{VC} \rho\beta\, d\forall$ es la **tasa de cambio de $\beta$ en el volumen de control $\mathrm{VC}$.**

- $\displaystyle \int_\mathrm{VC} \rho\beta\vec{V}_r\cdot\hat{n}dA$ representa el **flujo de $\beta$ a través de las fronteras del volumen de control ($\mathrm{VC}$)**

- **$\vec{V}_r$ es la *velocidad relativa* del flujo que cruza la frontera del volumen de control,** respecto de la velocidad de la frontera $\vec{V}_s$.

<img src="./figures/Reynolds_theorem_general.png" width="400px" align= center>

Notar que:

- para flujos de entrada $\vec{V}_r\cdot\hat{n} < 0$ ($\vec{V}_r$ apunta hacia dentro del volumen de control)
- para flujos de salida $\vec{V}_r\cdot\hat{n} > 0$ ($\vec{V}_r$ apunta hacia afuera del volumen de control)

Matemáticamente, el flujo neto de la propiedad $B$ en el volúmen de control:

\begin{equation*}
\dot{B}_\mathrm{neto} = \dot{B}_\mathrm{sale} - \dot{B}_\mathrm{entra} = \int_\mathrm{VC} \rho\beta\vec{V}\cdot\hat{n}dA
\end{equation*}

### Formulación simplificada

En la formulación anterior requiere conocer el comportamiento de los flujos en cada punto del dominio (para aplicar las integrales). Sin embargo, en muchas aplicaciones prácticas de ingeniería, solo se conocen valores promedios del flujo, obtenidos a través de mediciones (presión, caudal, etc).

Para estos casos, las siguientes aproximaciones son útiles:

- **Propiedades aproximadamente uniformes dentro del volúmen de control**

\begin{equation}
\frac{d}{dt}\int_\mathrm{VC} \rho\beta d\forall \approx \frac{d}{dt}(\rho\beta \forall)
\end{equation}

- **Valores promedio en la entrada y salida**

\begin{equation}\int_\mathrm{VC} \rho\beta\vec{V}\cdot\hat{n}dA \approx \sum_\mathrm{sale}\overline{\rho}\overline{\beta}~\overline{V}A - \sum_\mathrm{entra}\overline{\rho}\overline{\beta}~\overline{V}A
\end{equation}

donde $\bar{\beta}$ y $\overline{\rho}$ representan, respectivamente, el **valor promedio por área de la propiedad intensiva $\beta$ y de la densidad $\rho$**

\begin{equation}
\bar{\beta} = \frac{1}{A}\int_A \beta dA \quad\quad\quad
\bar{\rho} = \frac{1}{A}\int_A \rho dA,
\end{equation}

y $\overline{V}$ es la **velocidad promedio respecto al área $A$**

\begin{equation}
\overline{V} = \frac{1}{A}\int_A \vec{V}\cdot\hat{n}dA.
\end{equation}

<img src="./figures/average_velocity.png" width="250px" align= center>

A partir de estas aproximaciones deducimos la formulación simplificada del teorema de transporte de Reynolds:

\begin{equation}
\frac{d}{dt}B_\mathrm{sys} = \frac{d}{dt}(\rho\beta \forall) + \sum_\mathrm{sale}\overline{\rho}\overline{\beta}~\overline{V}A - \sum_\mathrm{entra}\overline{\rho}\overline{\beta}~\overline{V}A
\end{equation}

## Ecuación de conservación de masa

### Formulación general

En el caso de la conservación de masa del sistema

$$\frac{d}{dt}m_\mathrm{sys} = 0$$

Aplicando el teorema de transporte de Reynold con $B_\mathrm{sys} = m_\mathrm{sys}$ y $\beta = 1$, deducimos la ecuación de **conservación de masa para un volúmen de control**:

\begin{equation}
0 = \frac{d}{dt}\int_\mathrm{VC} \rho d\forall + \int_\mathrm{VC} \rho\vec{V}\cdot\hat{n}dA
\end{equation}

*Cualquier cambio en la masa dentro del volúmen de control ($ \frac{d}{dt}\int_\mathrm{VC} \rho d\forall$) se debe al flujo de masa hacia o desde el volúmen de control ($\int_\mathrm{VC} \rho\vec{V}\cdot\hat{n}dA$).*

Lidiamos con la ecuación de conservación de masa a menúdo en nuestra vida cotidiana

<img src="./figures/conservacion_masa_manguera.png" width="700px" align= center>

El ejemplo muestra un problema estacionario, donde $\frac{d}{dt} = 0$. Así, la ecuación de conservación de masa indica:

\begin{equation*}
0 = 0 + \rho V_2 A_2 - \rho V_1 A_1 \Rightarrow 
V_2 = V_1 \frac{A_1}{A_2}
\end{equation*}

Debido a que $A_1 > A_2$, entonces:

\begin{equation*}
V_2 > V_1
\end{equation*}

Explicando la salida con velocidad del agua al reducir el área de la manguera

### Variables relevantes

Aprovechamos esta ecuación para definir algunas variables relevantes:

#### Caudal ($Q$)
Se define como la **tasa de cambio del volúmen de fluido que cruza un área**, matemáticamente:

\begin{align}
Q &= \int_A \vec{V}\cdot\hat{n}dA&\quad\left(\mathrm{\frac{m^3}{s}}\right) &\quad\mathrm{(fórmula~general)}
\\
Q &=  \overline{V}A &\quad\left(\mathrm{\frac{m^3}{s}}\right) &\quad\mathrm{(valores~promedio)}
\end{align}

#### Flujo másico ($\dot{m}$)
Se define como la **tasa de cambio de la mása que cruza un área**, matemáticamente:

\begin{align}
\dot{m} &= \int_A \rho \vec{V}\cdot\hat{n}dA &\quad\left(\mathrm{\frac{kg}{s}}\right) &\quad\mathrm{(fórmula~general)}
\\
\dot{m} &= \overline{\rho} \overline{V}A = \rho Q &\quad\left(\mathrm{\frac{kg}{s}}\right) &\quad\mathrm{(valores~promedio)}
\end{align}

Si aplicamos esto considerando valores promedio, tenemos la siguiente expresión para la ecuación de conservación de masa:

\begin{equation}
0 = \frac{d}{d t} \rho \forall + \sum_\mathrm{sale}\dot{m} - \sum_\mathrm{entra}\dot{m}
\end{equation}

En el caso de un fluido incompresible:

\begin{equation}
0 = \frac{d}{d t} \forall + \sum_\mathrm{sale}Q - \sum_\mathrm{entra}Q
\end{equation}

## Ecuación de conservación de momentum lineal

### Formulación general

Según la ley de Newton, la **ecuación de conservación de momento lineal,** es:

\begin{equation*}
\frac{d}{dt}(m\vec{V})_\mathrm{sys} = \sum\vec{F}_\mathrm{ext}
\end{equation*}

donde $\vec{F}_\mathrm{ext}$ representa la fuerza externa neta sobre el sistema.

Aplicamos el teorema de transporte de Reynolds considerando: 
- Propiedad intensiva del momento lineal, $\beta = \vec{V}$
- Conservación de momento lineal aplicado al sistema $\frac{d}{dt}B_\mathrm{sys} = \frac{d}{dt}(m\vec{V})_\mathrm{sys} = \sum\vec{F}_\mathrm{ext}$.

La **ecuación de conservación de momento lineal para un V.C.** es:

\begin{equation}
\sum\vec{F}_\mathrm{ext} = \frac{d}{dt}\int_\mathrm{VC} \rho\vec{V} d\forall + \int_\mathrm{VC} \rho\vec{V}(\vec{V}_r\cdot\hat{n})dA
\end{equation}

*la suma de fuerzas externas aplicadas en un V.C. es igual a la tasa de cambio de momento lineal dentro del V.C. y al balance de momento lineal en la entrada y salida del V.C.*

>**Importante.** $\vec{V}$ es la velocidad absoluta del fluido, es decir, la velocidad del fluido respecto a un marco de referencia fijo (por ejemplo, el suelo).

La expresión para valores promedio es:

\begin{equation}
\sum \vec{F}_\mathrm{ext} = \frac{d}{d t} \rho \vec{V}\forall + \sum_\mathrm{sale}\dot{m}\vec{V} - \sum_\mathrm{entra}\dot{m}\vec{V}
\end{equation}

### Expresiones para volumen de control fijo y movil

La expresión anterior es una versión universal de la ecuación de conservación de momentum. Aunque la expresión es siempre efectiva, a veces, es más práctico considerar expresiones explícitas para un volumen de control fijo o móvil en base a la velocidad relativa del fluido respecto a la frontera del volumen de control, $\vec{V}_r$. Estas expresiones nacen de un análisis que considera la ecuación de conservación de masa.

**Volumen de control fijo.** La velocidad relativa del fluido respecto a la frontera del volumen de control es igual a la velocidad absoluta del fluido, es decir, $\vec{V}_r = \vec{V}$. Por lo tanto, la ecuación de conservación de momento lineal se simplifica a:

\begin{equation}
\sum\vec{F}_\mathrm{ext} = \frac{d}{dt}\int_\mathrm{VC} \rho\vec{V} d\forall + \int_\mathrm{VC} \rho\vec{V}(\vec{V}\cdot\hat{n})dA
\end{equation}

**Volumen de control móvil.** La velocidad relativa del fluido respecto a la frontera del volumen de control es diferente a la velocidad absoluta del fluido. En este caso, la expresión relativa a $\vec{V}_r$ debe considerar la aceleración del volumen de control, $\vec{a}_\mathrm{VC}$, respecto a un marco de referencia inercial.

\begin{equation}
\sum\vec{F}_\mathrm{ext} - m_\mathrm{VC}\frac{d\vec{V}_\mathrm{VC}}{dt} = \frac{d}{dt}\int_\mathrm{VC} \rho\vec{V}_r d\forall + \int_\mathrm{VC} \rho\vec{V}_r(\vec{V}_r\cdot\hat{n})dA
\end{equation} 

### Fuerzas externas $\vec{F}_\mathrm{ext}$

Las fuerzas externas $\sum\vec{F}_\mathrm{ext}$ sobre un volumen de control se deben a:
1. Fuerzas que aparecen en el corte de cuerpos sólidos que penetran a través de la superficie de control
2. Fuerzas debidas a presión y viscosidad en el fluido del contorno.
3. Fuerzas de cuerpo, como la gravedad ($\rho \vec{g}\forall$), fuerzas eléctricas, etc.

**En el caso de fuerzas de superficie, en esta unidad nos centraremos en las fuerzas de presión**, aunque en unidades posteriores revisaremos también las fuerzas de viscosidad.

#### Fuerzas de presión
La fuerza de presión sobre una superficie es perpendicular a ésta y dirigida hacia ella.

<img src="./figures/fuerza_neta_presion.png" width="700px" align= center>

Si la presión tiene un valor uniforme, la resultante es nula. Por ejemplo, en la figura de la izquierda:

\begin{equation}
F_p = \int_S p_a (-\hat{n})dA = -p_a\int_S \hat{n}dA = 0
\end{equation}

Si no, la resultante tomará relevancia en las regiones con diferencia de presiones. Por ejemplo, en el caso de la figura derecha:

\begin{equation}
F_p = \int_S p (-\hat{n})dA = \sum_i p_{i, \mathrm{man}}A_i (-\hat{n}_i)
\end{equation}

donde $A_i$ y $\hat{n}_i$ corrseponden, respectivamente, al área y vector normal (apuntando hacia afuera del volumen) de las superficies con presión diferencia de presiones.

## Ecuación de conservación de momentum angular

### Formulación general

Según la ley de Newton, la **ecuación de conservación de momento angular,** respecto a un eje de referencia inercial en $O$ es:

\begin{equation*}
\frac{d}{dt}\vec{L}_\mathrm{sys} = \sum\vec{M}_{O, \mathrm{ext}}
\end{equation*}

donde $\vec{M}_{O, \mathrm{ext}} = (\vec{r}_O \times \vec{F}_{\mathrm{ext}})$ representa el momento externo neto sobre el sistema respecto un punto $O$.

Aplicamos el teorema de transporte de Reynolds considerando: 
- Propiedad intensiva del momento angular, $\beta = \vec{r} \times \vec{V}$
- Conservación de momento angular aplicado al sistema $\frac{d}{dt}B_\mathrm{sys} = \frac{d}{dt}\vec{L}_\mathrm{sys} = \sum\vec{M}_O$.

La **ecuación de conservación de momento angular para un V.C.** es:

\begin{equation}
\sum\vec{M}_{O, \mathrm{ext}} = \frac{d}{dt}\int_\mathrm{VC} \rho(\vec{r}\times\vec{V}) d\forall + \int_\mathrm{VC} \rho(\vec{r}\times\vec{V})(\vec{V}_r\cdot\hat{n})dA
\end{equation}

*la suma de momentos externos aplicados en un V.C. es igual a la tasa de cambio de momento angular dentro del V.C. y al balance de momento angular en la entrada y salida del V.C.*

>**Importante.** Similar a la ecuación de conservación de momento lineal, $\vec{r}\times\vec{V}$ representa el momento angular respecto a un eje inercial de referencia (por ejemplo, el suelo).

Si aplicamos esto considerando valores promedio, tenemos la siguiente expresión para la ecuación de conservación de masa:

\begin{equation}
\sum \vec{M}_{O, \mathrm{ext}} = \frac{d}{d t} \rho\forall (\vec{r} \times \vec{V}) + \sum_\mathrm{sale}\dot{m}(\vec{r} \times \vec{V}) - \sum_\mathrm{entra}\dot{m}(\vec{r} \times \vec{V})
\end{equation}

### Torque externo neto $\vec{M}_O$
Al igual que el caso de momento lineal, el torque externo neto $\vec{M}_O$ resulta de la acción de las fuerzas externas sobre el sistema, y se puede expresar como:
\begin{equation}
\vec{M}_O = \vec{r}\times \vec{F}_s +  \vec{r}_\mathrm{G} \times m\vec{g} + \vec{T}_\mathrm{eje}
\end{equation}
donde:
- $\vec{r}\times \vec{F}_s$ es el torque debido a fuerzas de superficie (por ejemplo, fuerzas de presión o interacción con cuerpos sólidos)
- $\vec{r}_\mathrm{G} \times m\vec{g}$ es el torque debido a fuerzas de gravedad, con $\vec{r}_\mathrm{G}$ el vector de posición del centro de gravedad del sistema respecto al eje $O$.
- $\vec{T}_\mathrm{eje}$ es el torque debido a fuerzas de fricción en el eje de rotación (por ejemplo, en una turbina hidráulica)

### Análisis para maquinas rotativas
En el caso de máquinas rotativas, como turbinas hidráulicas o aerogeneradores, es posible simplificar la ecuación de conservación de momento angular asumiendo:
- Volumen de control fijo, centrado en el eje de rotación
- Centro de masa coincide con el eje de rotación, es decir, $\vec{r}_\mathrm{G} = 0$.
- Fuerzas de superficie uniformes alrededor de la superficie, es decir $\vec{r}\times \vec{F}_s = 0$

En este caso, la ecuación de conservación de momento angular se simplifica a:

\begin{equation}
\vec{T}_\mathrm{eje} - I\frac{d\omega}{dt}= \sum_\mathrm{sale}\dot{m}(\vec{r} \times \vec{V}) - \sum_\mathrm{entra}\dot{m}(\vec{r} \times \vec{V})
\end{equation}

donde $I$ es el momento de inercia del rotor respecto al eje de rotación, y $\omega$ es la velocidad angular del rotor.

## Ecuación de conservación de energía

### Formulación general

A partir del teorema de transporte de Reynolds, y considerando la propiedad intensiva de la energía, $\tilde{e}$, definida como la **energía por unidad de masa** ($\mathrm{kJ/kg}$, por ejemplo), tenemos que la **ecuación de conservación de energía aplicada a un volúmen de control** es:

\begin{equation}
\dot{Q} - \dot{W} = \frac{d}{dt}\int_\mathrm{VC} \rho\tilde{e}d\forall + \int_\mathrm{VC} \rho\tilde{e}\vec{V}\cdot\hat{n}dA,\quad(\mathrm{kW})
\end{equation}

donde:
- $\dot{Q}$, tasa de transferecia calor recibida por el sistema ($\mathrm{kW}$)
- $\dot{W}$, tasa de trabajo (potencia) entregada por el sistema ($\mathrm{kW}$)

#### Energía del fluido ($\tilde{e}$)

Considera todas las formas de energía presentes en el fluido:

\begin{equation*}
\tilde{e} = \tilde{e}_\mathrm{interna} + \tilde{e}_\mathrm{cinética} + \tilde{e}_\mathrm{potencial} + \cdots, \quad\left(\mathrm{\frac{kJ}{kg}}\right)
\end{equation*}

donde:
- $\tilde{e}_\mathrm{interna} = \tilde{u}$ (energía interna)
- $\tilde{e}_\mathrm{cinética} = \frac{V^2}{2}$ (energía cinética)
- $\tilde{e}_\mathrm{potencial} = gz$ (energía potencial)

> $V$ es la velocidad promedio del fluido, $z$ es la altura medida desde un punto de referencia común, y $g$ es la aceleración de gravedad.

#### Transferencia de calor ($\dot{Q}$)

Energía asociada al calor transferido por diferencia de temperaturas. 
>Notar que $\dot{Q} > 0$ si el volúmen de control recibe calor

#### Transferencia de trabajo ($\dot{W}$)

Considera el trabajo entregado o recibido por el sistema.
>Notar que $\dot{W} > 0$ si el sistema entrega trabajo.

Distinguimos tres formas principales de trabajo:

- **Trabajo de eje ($\dot{W}_\mathrm{eje}$),** asociado al trabajo intercambiado con maquinarias, por ejemplo, trabajo de bombas ($\dot{W}_\mathrm{eje} < 0$), turbinas ($\dot{W}_\mathrm{eje} > 0$), ventiladores ($\dot{W}_\mathrm{eje} < 0$), etc.


- **Trabajo por presiones ($\dot{W}_\mathrm{presión}$),** asociado al trabajo de presiones del fluido. Actúa solo en las fronteras del volúmen de control:

    \begin{equation*}
    \dot{W}_\mathrm{presión} = \int_{VC} p(\vec{V}\cdot\hat{n})dA
    \end{equation*}

 - **Trabajo por fuerzas viscosas ($\dot{W}_\mathrm{viscosas}$),** asociado a la disipación de energía por fricción como resultado de viscosidad del fluido ($\dot{W}_\mathrm{viscosas}$ > 0).

### Ecuación de conservación de energía mecánica

La ecuación de conservación de energía (3.14) no tiene mucho uso práctico en la ingeniería, debido a que los trabajos asociados a la presión y viscosidad están implícitos en la ecuación. 

Una formulación más común se obtiene separando los trabajos de eje, presión y viscosas:

<div class="full-width">
<p style="line-height:2.0em;"></p>
    
\begin{equation*}
\dot{Q} - \dot{W}_\mathrm{eje} - \dot{W}_\mathrm{viscoso}  = \frac{d}{dt}\int_\mathrm{VC} \rho\left(\tilde{u}+\frac{V^2}{2}+gz\right)d\forall + \int_\mathrm{VC} \rho\left(\tilde{h}+\frac{V^2}{2}+gz\right)\vec{V}\cdot\hat{n}dA,\quad(\mathrm{kW})
\end{equation*}
    
 </div>
    
donde $\tilde{h} = \tilde{u} + p/\rho$ es la **entalpía o *energía de flujo* del fluido.**

Si el sistema analizado no está sujeto a efectos térmicos (problema isotérmico y adiabático), $\dot{Q} = 0$ y $\tilde{u} = cte$, y obtenemos la **ecuación de conservación de la energía mecánica**:

<div class="full-width">
<p style="line-height:2.0em;"></p>
    
\begin{equation*}
- \dot{W}_\mathrm{eje} - \dot{W}_\mathrm{viscoso}  = \frac{d}{dt}\int_\mathrm{VC} \rho\left(\frac{V^2}{2}+gz\right)d\forall + \int_\mathrm{VC} \rho\left(\frac{P}{\rho}+\frac{V^2}{2}+gz\right)\vec{V}\cdot\hat{n}dA,\quad(\mathrm{kW})
\end{equation*}
    
</div>

Considerando un **problema estacionario, con valores promedios en las entradas y salidas del volúmen de control**:

<div class="full-width">
<p style="line-height:2.0em;"></p>
    
\begin{equation*}
- \dot{W}_\mathrm{eje} - \dot{W}_\mathrm{viscoso}  = \sum_\mathrm{sale}\left(\frac{P}{\rho}+\frac{V^2}{2}+gz\right)\dot{m} - \sum_\mathrm{entra}\left(\frac{P}{\rho}+\frac{V^2}{2}+gz\right)\dot{m},\quad(\mathrm{kW})
\end{equation*}
    
</div>

### Ecuación de conservación de energía mecánica para una línea de flujo

En base a los casos anteriores, y considerando **(1) flujo incompresible y (2) volúmen de control con una única entrada y salida**, tenemos:

\begin{equation}
\left(\frac{P}{\rho g}+\frac{V^2}{2g}+z\right)_\mathrm{entra} + h_\mathrm{bomba} = \left(\frac{P}{\rho g}+\frac{V^2}{2g}+z\right)_\mathrm{sale} + h_\mathrm{turbina} + h_L,\quad(\mathrm{m})
\end{equation}

donde $h_\mathrm{i} = \frac{\dot{W}_\mathrm{i}}{\rho gVA}$, ($i=$ $\mathrm{bomba}$, $\mathrm{turbina}$ y $L$) es la **altura de presión equivalente** asociada a la energía transferida por una bomba, turbina o viscosidad, respectivamente.

El término $h_\mathrm{L}$ se conoce como la **pérdida de carga,** asociada al pérdida irreversible de energía mecánica por fricción y viscosidad.

La ecuación (4.7) representa la forma más utilizada de la ecuación de conservación energía mecánica. Se expresa en forma de **alturas hidrostáticas,** ya que permite identificar los cambios de presión en el ducto facilmente.

### Ecuación de Bernoulli

Si aplicamos la ecuación anterior en un tramo de un ducto, donde $h_\mathrm{bomba} = h_\mathrm{turbina} = 0$, y asumimos **fluido ideal, es decir, sin fricciones por viscosidad ($h_L = 0$)**, tenemos la **ecuación de Bernoulli.**

\begin{equation}
\left(\frac{P}{\rho}+\frac{V^2}{2}+gz\right)_\mathrm{sale} = \left(\frac{P}{\rho}+\frac{V^2}{2}+gz\right)_\mathrm{entra}
\end{equation}

Esta ecuación, aunque tiene poco uso práctico, permite comprender en términos simples la conservación de energía en un fluido producto de los cambios de presión, velocidad y elevación.
